```python
#TWEEN 80 1WT%
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
label1 = 'Tween 80 1wt%'
y = np.array([46.26783953, 45., 41.34283304, 42.76991441, 39.39929111, 41.95215191])
x = np.array([1, 5, 10, 15, 20, 25])
e = np.array([0.723524522, 0, 0.317314567, 0.603928941, 0.202296533, 0.546387435])

plt.figure ("Surface Tension v.s Temperature" , figsize = [8,5])
plt.title ("Surface Tension v.s Temperature " , fontsize = 13)
plt.xlabel(r'Temperature [$\degree C$]', fontsize = 11) 
plt.ylabel(r'Tension [mN]',fontsize = 11)
plt.errorbar(x, y, e, linestyle='None', marker='^')


plt.show()
```

```python
#Tension v.s Concentration
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
label3 = 'Tween 80 - 1$\degree C$'
label4=  'Tween 20 - 1$\degree C$'
y_t80 = np.array([54.48675579, 48.40625711, 46.26783953, 45.36468188])
x_t80 = np.array([0.01, 0.1, 1, 2])
e_t80 = np.array([0.873719803, 0.938253271, 0.723524522, 0.669043576])

y_t20 = np.array([41.67341651, 42.4483062, 44.33043616, 53.88520901])
x_t20 = np.array([2, 1 ,0.1,0.01])
e_t20 = np.array([0.875600962, 0.685660558, 1.113147372, 1.067864029 ])

plt.figure ("Surface Tension v.s Concentration" , figsize = [8,5])
plt.title ("Surface Tension v.s Concentration " , fontsize = 13)
plt.xlabel(r'Conc. [wt. %]', fontsize = 11) 
plt.ylabel(r'Tension [mN]',fontsize = 11)
plt.errorbar(x_t80, y_t80, e_t80, linestyle='None', marker='^')
plt.errorbar(x_t20, y_t20, e_t20, linestyle='None', marker='^')
plt.legend([label3,label4])

plt.show()
```
