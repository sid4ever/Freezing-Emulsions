```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline


plt.figure(dpi=150)

p1 = plt.plot([1,2,3,4,5], [5,5,5,5,5], 'ro') #Planar front plotting

p2 = plt.plot([6,7,8], [5,10,5],linestyle='',marker= 'o',markersize='6',color ='red', markerfacecoloralt='black',fillstyle='left')

p3 = plt.plot([9,10],[5,10],'ko') #Instabilities

p4 = plt.plot([1,2,3,4,5], [10,10,10,10,10], 'ro') #planar front plotting

plt.xticks(np.arange(0,11,1))
plt.yticks(np.arange(0,16,5))

plt.xlabel('Interface Velocity ' r' $(\mu$m/sec)')
plt.ylabel('Temperature Gradient 'r' $(\degree$C/mm)')

plt.title('Destabilisation')

plt.grid(True,which='major',axis='x',color='grey', alpha=0.2)
plt.grid(True,which='major',axis='y',color='green', alpha=0.2)

plt.legend((p1[0],p2[0],p3[0]),('Planar','No Instability','Instabilities'))


plt.savefig("Destablisation Phase Diagram", dpi=300)
```
